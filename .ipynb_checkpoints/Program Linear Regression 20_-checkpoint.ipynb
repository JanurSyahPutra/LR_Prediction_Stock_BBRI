```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

plt.style.use('seaborn-darkgrid')
plt.rc('figure', figsize=(13,9))
```

```python
dataset = pd.read_csv("BBRI Historical Data.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset['Price'] = dataset['Price'].str.replace(',', '').astype(int)
dataset['Open'] = dataset['Open'].str.replace(',', '').astype(int)
dataset['High'] = dataset['High'].str.replace(',', '').astype(int)
dataset['Low'] = dataset['Low'].str.replace(',', '').astype(int)
dataset['Vol.'] = dataset['Vol.'].str.replace('M', '').astype(float)
dataset['Change %'] = dataset['Change %'].str.replace('%', '').astype(float)
```

```python
dataset.info()
```

```python
dataset = dataset.set_index('Date')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset = pd.DataFrame(dataset, columns=['Price'])
dataset = dataset.reset_index()
```

```python
dataset
```

```python
dataset.isna().values.any()
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots()
ax.plot(dataset['Date'], dataset['Price'])

plt.title('Harga Penutupan Saham pada Bank BRI Periode 2008 sampai 2020', fontsize=16)
plt.xlabel('Tahun / Periode', fontsize=14)
plt.ylabel('Harga Penutupan Saham (Rp)', fontsize=14)

fig.autofmt_xdate()
plt.show()
```

```python
train, test = train_test_split(dataset, test_size=0.20, random_state=42)

X_train = np.array(train.index).reshape(-1, 1)
y_train = train['Price']

X_test = np.array(test.index).reshape(-1, 1)
y_test = test['Price']
```

```python
print('Jumlah Data Latih',train.shape)
```

```python
print('Jumlah Data Uji',test.shape)
```

```python
LR = LinearRegression()

LR.fit(X_train, y_train)
```

```python
LR.score(X_train, y_train)
```

```python
LR.score(X_test, y_test)
```

```python
X = np.array(train.index).reshape(-1, 1)
Y = train['Price']

a = LR.intercept_
b = LR.coef_
Y_pred = a + b*X

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')
plt.title('Harga Penutupan Saham pada Bank BRI Periode 2008 sampai 2020', fontsize=16)
plt.xlabel('X (Date (Integer))', fontsize=14)
plt.ylabel('Y (Harga Penutupan Saham)', fontsize=14)
fig.autofmt_xdate()
plt.show()

print('Intercept : ', a)
print('Slope : ', b)
```

```python
y_pred_train = LR.predict(X_train)
y_pred = LR.predict(X_test)
```

```python
#Menguji Prediksi Menggunakan Data train dengan data prediksi (train)
plt.title('Prediksi Linear Regression | Date dan Price (Latih)', fontsize=16)
plt.scatter(X_train, y_train, label='Harga Prediksi')
plt.scatter(X_train, y_pred_train, label='Harga Aktual')

plt.xlabel('Date (integer)', fontsize=14)
plt.ylabel('Price (Rp)', fontsize=14)

plt.legend()
plt.show()
```

```python
train_dt = train
train['Prediction'] = y_pred_train
train.to_excel('Train_prediksi.xlsx')
train.shape
```

```python
#Menguji Prediksi Menggunakan Data test dengan data prediksi (test)
plt.title('Prediksi Linear Regression | Date dan Price (Uji)', fontsize=16)
plt.scatter(X_test, y_pred,  label='Harga Prediksi')
plt.scatter(X_test, y_test,  label='Harga Aktual')

plt.xlabel('Date (integer)', fontsize=14)
plt.ylabel('Price (Rp)', fontsize=14)

plt.legend()
plt.show()
```

```python
test_dt = test
test['Prediction'] = y_pred
test.shape
```
