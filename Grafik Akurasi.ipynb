```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style="ticks")
plt.figure(figsize=(12, 6))

x = np.array(["60/40", "65/35", "70/30", "75/25", "80/20"])
train = np.array([0.8977, 0.8977, 0.8992, 0.8988, 0.8992])
test = np.array([0.9078, 0.9086, 0.9073, 0.9101, 0.9125])

plt.title("Perbandingan Akurasi Berdasarkan Rasio Train dan Test", fontsize=16)
plt.xlabel("Rasio Pembagian Train dan Test", fontsize=15)
plt.ylabel("Akurasi", fontsize=15)

plt.plot(x, train, label = "Train", marker='o')
plt.plot(x, test, label = "Test", marker='o')
plt.legend()
plt.grid(color = 'black', linestyle = '-', linewidth = 1)
plt.show()
```
